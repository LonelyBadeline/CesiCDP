    <h3>Représentation des Données</h3>
    <ul>
        <li><h4>Graphe</h4></li>
        Un graphe est une structure composée d'objets avec des relations entre ces derniers.
        Les objets sont définis comme des  « sommets » (ou nœuds/points) la ou les relations portent le nom d’ « arêtes » (ou liens ou lignes). On utilise les sommets pour représenter des entités (ex : villes), tandis que les relations servent à représenter des connexion entre nos entités (ex : routes).
        Un graphe se définit comme un couple G=(V,E) :
                        V est l’ensemble des sommets
        E est l’ensemble des arêtes, associé de manière individuel à un couple (u,v) ou une paire de sommets (u, v  V).
                        Il existe différent type de graphes :
                        Les graphes orientés :  les arêtes ont une direction définie.
                        Les graphes non orientés : les arêtes sont bidirectionnelles.
                        Les graphes pondérés : les arêtes sont chacune associé à un poid (une valeur).
        <br>
        <br>
        <br>
        <li><h4>Calcul Matriciel</h4></li>
            On appelle matrice d'adjacence associée à ce graphe la matrice A dont le terme aij vaut 1 si les sommets i et j sont reliés par une arête et 0 sinon.
            <br>
            Dans le cas d'un graphe non orienté, les coefficients aij et aji sont égaux pour tout i et tout j compris entre 1 et n. Autrement dit, la matrice d'adjacence est symétrique.
            Dans le cas d'un graphe orienté, la matrice d'adjacence n'est pas a priori symétrique
            Il existe également une variante de la matrice d’adjacence nommée matrice de poids où les valeurs indique la longueur de l’arête entre deux sommets. Si il n’existe pas de lien on utilise ∞ pour le représenter.
            <br>
            Source : <a href="https://math.univ-lyon1.fr/irem/Formation_ISN/formation_parcours_graphes/graphe/2_matrice.html" target="_blank">Matrice d'adjacence Univ Lyon</a>
    </ul>
